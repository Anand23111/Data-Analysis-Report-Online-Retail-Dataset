```python
#q5.Compute the mean vector  for the data matrix, and then compute the total variance 
```

```python
data = [
    [6, 7],
    [2, 8],
    [12,34],
    
]

# Convert the data list into a numpy array
data_matrix = np.array(data)

# Print the first few rows of the data matrix
print(data_matrix[:10])

#  Calculate the mean vector 
mean_vector = np.mean(data_matrix, axis=0)
print("mean vector: ",mean_vector)
print("-----------")
centered_data = data_matrix - mean_vector
print(centered_data)
print("----------")

total_variance = np.sum(centered_data ** 2) / (data_matrix.shape[0] * data_matrix.shape[1])


print("Total Variance:", total_variance)
```

```python
#Q6.

centered_data_matrix = np.array([
    [6, 7],
    [2, 8],
    [12,34],
])


inner_product_covariance = np.dot(centered_data_matrix.T, centered_data_matrix) / (centered_data_matrix.shape[0] - 1)


outer_product_covariance = np.cov(centered_data_matrix, rowvar=False)

print("Sample Covariance Matrix (Inner Product):")
print(inner_product_covariance)

print("\nSample Covariance Matrix (Outer Product):")
print(outer_product_covariance)
```

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt




print(data.head())


X = data.drop('Quantity', axis=1)  

# apply PCA 
pca = PCA(n_components=2)  
X_pca = pca.fit_transform(X)

# apply t-sne 
tsne = TSNE(n_components=2, perplexity=30)  
X_tsne = tsne.fit_transform(X)

# Plot the results
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=data['Quantity'], cmap='viridis')
plt.title('PCA Visualization')
plt.colorbar()

plt.subplot(1, 2, 2)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=data['UnitPrice'], cmap='viridis')
plt.title('t-SNE Visualization')
plt.colorbar()

plt.tight_layout()
plt.show()
```

```python
#PCA Visualization:

#In a PCA plot, the first component (x-axis) captures 
#the most variance in the data, followed by the second component (y-axis).
#If you see distinct clusters or separation of data points in the PCA visualization,
#it suggests that the data might have inherent patterns or groupings. 
#However, keep in mind that PCA focuses on preserving maximum variance, which might not always lead to the most meaningful separations.

#Clusters: If data points from the same class or group
#are tightly clustered together, while different groups are
#well-separated, it indicates that the original high-dimensional data has certain distinguishable patterns.

#Spread: If the data points are spread out without clear clustering,
#it might indicate that the data is more continuous 
#or doesn't exhibit clear separations in lower dimensions.
```
